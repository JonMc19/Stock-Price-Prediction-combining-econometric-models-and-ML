```python
api_key = "6893185bee4333.79098361" #EODHD API key

ticker= "AAPL.US"  # Example ticker for Apple Inc.

url = f"https://eodhd.com/api/fundamentals/{ticker}?api_token=demo&fmt=json"

response = requests.get(url)
data = response.json()
```

```python
income_statement = data["Financials"].get("Income_Statement", {})
yearly = income_statement.get("yearly", {})

if not yearly:
    print("No yearly income statement data found.")
else:
    # Extract years from date keys (like '2024-09-30')
    years = [date_str[:4] for date_str in yearly.keys()]
    max_year = max(years)

    print(f"Max year available: {max_year}\n")
    print("Yearly Revenue and Net Income:")

    # For each year, print revenue and net income
    # Note: 'revenue' field might not exist, if so try 'incomeBeforeTax' or 'grossProfit'
    for date_str in sorted(yearly.keys(), reverse=True):
        year = date_str[:4]
        data_year = yearly[date_str]
        revenue = data_year.get("revenue") or data_year.get("incomeBeforeTax") or data_year.get("grossProfit") or "N/A"
        net_income = data_year.get("netIncome", "N/A")
        print(f"Year {year}: Revenue = {revenue}, Net Income = {net_income}")
```
