```python
import pandas as pd
```

```python
fundamentals = pd.read_csv('/Users/jon/Desktop/Ironhack/Unit 7 - Final Project/Final-Project-Ironhack/Type of analysis/Time series analysis/fundamentals/csv/adobe/quarterly/adobe_quarterly_all_metrics.csv')
macros = pd.read_csv('/Users/jon/Desktop/Ironhack/Unit 7 - Final Project/Final-Project-Ironhack/Type of analysis/Time series analysis/macro/csv/macro_2009_2025.csv')
prices = pd.read_csv('/Users/jon/Desktop/Ironhack/Unit 7 - Final Project/Final-Project-Ironhack/Type of analysis/Time series analysis/prices/csv/adbee_price.csv')
```

```python
macros = macros.sort_values(by='date', ascending=True)
macros = macros.reset_index(drop=True)
macros
```

```python
prices
```

BEFORE MERGING MAKE SURE DATE COLUMN BETWEEN MACRO AND FUNDAMENTAL DATA CORRESPOND IF NOT THE DF WILL COME OUT EMPTY

```python
#If dates on fundamentals coincide with macros, uncomment line 3, otherwise, use line 4 and 5. May have to adjust depending on the dataset.

#stock = pd.merge(funadamentals, macros, on='date', how='inner')
stock = pd.merge(fundamentals, macros, left_index=True, right_index=True, how="inner")
stock = stock.rename(columns={'date_x': 'date'})


stock
```

```python
stock = pd.merge(stock, prices, left_on='date', right_on='Date', how='inner')
stock.drop(columns=['Date', 'Unnamed: 0', 'calendar_quarter'], inplace=True)  # Drop redundant columns if necessary
# Reorder columns to have 'date' first and Price next to it
cols = stock.columns.tolist()
cols.insert(1, cols.pop(cols.index('Price')))
stock = stock[cols]
stock
```

```python
stock.to_csv('adbe_data.csv')
```
