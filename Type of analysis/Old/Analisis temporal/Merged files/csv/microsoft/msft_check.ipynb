```python
import pandas as pd
```

```python
msft_data=pd.read_csv("/Users/jon/Desktop/Ironhack/Unit 7 - Final Project/Final-Project-Ironhack/Type of analysis/Analisis temporal/Merged files/csv/microsoft/merged_microsoft_funadmentals_price_macro.csv")
```

```python
msft_data
```

```python
#Calculate average revenue growth rate
msft_data['Revenue Growth Rate'] = msft_data['revenue'].pct_change() * 100
avg_revenue_growth_rate = msft_data['Revenue Growth Rate'].mean()
print(f"Average Revenue Growth Rate: {avg_revenue_growth_rate:.2f}%")

#Calculate average net income growth rate
msft_data['Net Income Growth Rate'] = msft_data['net_income'].pct_change() * 100
avg_net_income_growth_rate = msft_data['Net Income Growth Rate'].mean()
print(f"Average Net Income Growth Rate: {avg_net_income_growth_rate:.2f}%")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(msft_data['period_end'], msft_data['revenue'], label='Revenue', color='blue')
plt.plot(msft_data['period_end'], msft_data['net_income'], label='Net Income', color='green')

plt.xlabel('Period End')
plt.ylabel('Amount')
plt.title('Revenue and Net Income Over Time')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
