```python
import pandas as pd
```

```python
funadamentals = pd.read_csv('/Users/jon/Desktop/Ironhack/Unit 7 - Final Project/Final-Project-Ironhack/Type of analysis/Analisis temporal/fundamentals/csv/nike/quarterly/nike_quarterly_all_metrics.csv')
macros = pd.read_csv('/Users/jon/Desktop/Ironhack/Unit 7 - Final Project/Final-Project-Ironhack/Type of analysis/Analisis temporal/macro/csv/macro_2009_2025.csv')
prices = pd.read_csv('/Users/jon/Desktop/Ironhack/Unit 7 - Final Project/Final-Project-Ironhack/Type of analysis/Analisis temporal/prices/python/nike/nike_price.csv')
```

```python
funadamentals['date'] = pd.to_datetime(funadamentals['date'].str.strip())
macros['date'] = pd.to_datetime(macros['date'].str.strip())
prices['Date'] = pd.to_datetime(prices['Date'].str.strip())
```

```python
funadamentals
```

```python
macros
```

```python
#When companies don't report earnings during the calendar quarter we have to forward fill the data
macros = macros.set_index('date').resample('ME').ffill().reset_index()
```

```python
prices
```

```python
microsoft = pd.merge(funadamentals, macros, on='date', how='inner')
microsoft = pd.merge(microsoft, prices, left_on='date', right_on='Date', how='inner')
microsoft.drop(columns=['Date', 'Unnamed: 0', 'calendar_quarter'], inplace=True)  # Drop redundant columns if necessary

# Reorder columns to have 'date' first and Price next to it
cols = microsoft.columns.tolist()
cols.insert(1, cols.pop(cols.index('Price')))
microsoft = microsoft[cols]
microsoft
```

```python
microsoft.to_csv('nike_data.csv')
```
