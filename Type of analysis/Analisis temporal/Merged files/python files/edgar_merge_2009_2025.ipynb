```python
import pandas as pd
```

```python
financials_09_14 = pd.read_csv("/Users/jon/Desktop/Ironhack/Unit 7 - Final Project/Final-Project-Ironhack/Type of analysis/Analisis temporal/Fundamentals/Data/csv files/own code/apple_financials_2009_2014.csv")
financials_14_25 = pd.read_csv("/Users/jon/Desktop/Ironhack/Unit 7 - Final Project/Final-Project-Ironhack/Type of analysis/Analisis temporal/Fundamentals/Data/csv files/own code/apple_financials_2014_2025.csv")
```

```python
# Add a marker: before 2014 → already in full dollars
financials_09_14["scale"] = 1
# After 2014 → reported in thousands (need ×1000)
financials_14_25["scale"] = 1000

# Concatenate
financials_09_25 = pd.concat([financials_09_14, financials_14_25], ignore_index=True)

# Normalize all numeric columns to dollars
num_cols = ["revenue","net_income","total_assets","total_liabilities","shareholders_equity"]
for c in num_cols:
    financials_09_25[c] = financials_09_25[c] * financials_09_25["scale"]

# Drop helper column
financials_09_25 = financials_09_25.drop(columns=["scale", "xbrl_source"])

# Sort by filing date
financials_09_25["period_end"] = pd.to_datetime(financials_09_25["period_end"])
financials_09_25 = financials_09_25.sort_values("period_end")

# Save unified dataset
financials_09_25.to_csv("apple_financials_2009_2025.csv", index=False)

print("✅ Saved apple_financials_2009_2025.csv with consistent units (full dollars)")
```
