```python
import pandas as pd
```

```python
fundamentals=pd.read_csv("/Users/jon/Desktop/Ironhack/Unit 7 - Final Project/Final-Project-Ironhack/Type of analysis/Analisis temporal/Merged files/csv/apple_financials_Clean_09_25_complete.csv")
prices=pd.read_csv("/Users/jon/Desktop/Ironhack/Unit 7 - Final Project/Final-Project-Ironhack/Type of analysis/Analisis temporal/Prices/csv/apple_price.csv")
gdp=pd.read_csv("/Users/jon/Desktop/Ironhack/Unit 7 - Final Project/Final-Project-Ironhack/Type of analysis/Analisis temporal/Macroeconomic/csv files/gdp_2010_2025.csv")
```

```python
#--FUNDAMENTALS DF--

#Drop unecessary columns for fundamentals df - update if needed
fundamentals=fundamentals.drop(columns=["period_end", "form", "fq", "fy"])

#Create ratio assets/liabilities
fundamentals["ratio assets/libailities"] = fundamentals["total_assets"] / fundamentals["total_liabilities"]

#Drop assets and liabilities columns
fundamentals = fundamentals.drop(columns=["total_assets", "total_liabilities"])
```

```python
#--PRICES DF--

#Keep only prices for the last day of each quarter

# ensure it's datetime
prices["Date"] = pd.to_datetime(prices["Date"], utc=True).dt.tz_convert(None)

# keep only rows where day/month match your condition
prices = prices[prices["Date"].dt.strftime("%d/%m").isin(["01/04", "01/07", "01/10", "01/01"])]

# keep only rows from 2010 onwards
prices = prices[prices["Date"].dt.year >= 2010]

#Add quareter column
prices["quarter"] = "Q" + prices["Date"].dt.quarter.astype(str)+" "+ prices["Date"].dt.year.astype(str)

#remove uncessary columns for prices df - update if needed
prices = prices.drop(columns=["Date", "High", "Low", "Close", "Volume", "Dividends", "Stock Splits"])

#Change column name -> Open to Price
prices.rename(columns={"Open": "Price"}, inplace=True)
```

```python
#--GDP DF--

gdp.rename(columns={"Value":"gdp_growth"}, inplace=True)
gdp=gdp.drop(columns=["Date"])

```

```python
#merge funadmentals and prices dataframes
merged = pd.merge(fundamentals, prices, on="quarter", how="left")

#merge gdo

merged = merged.merge(gdp, on="quarter", how="left")

#Reorder columns
merged = merged[["quarter","Price","revenue", "net_income", "ratio assets/libailities", "shareholders_equity", "gdp_growth"]]

merged.to_csv("apple_funadmentals_price_gdp.csv", index=False)
```
