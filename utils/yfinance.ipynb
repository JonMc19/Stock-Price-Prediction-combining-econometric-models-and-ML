```python
import yfinance as yf # Python library that lets you download stock market data from Yahoo Finance
import pandas as pd
```

```python
#Set the ticker symbol for the stock you want to analyze
ticker = yf.Ticker("AAPL")

#Check the metrics available for the ticker
income_stmt = ticker.financials.T  # transpose for chronological order

print(income_stmt.columns)  # inspect all available fields
```

```python
#Company Info

info = ticker.info
print("Sector:", info.get('sector', 'N/A'))
print("Industry:", info.get('industry', 'N/A'))
print("PE ratio:", info.get('trailingPE', 'N/A'))
print("Market Cap:", info.get('marketCap', 'N/A'))
print("Country:", info.get('country', 'N/A'))

# Collect Financial Statements (basic)

income_stmt = ticker.financials
balance_sheet = ticker.balance_sheet
cashflow = ticker.cashflow

#print("\nIncome Statement:")
#print(income_stmt.head())
#print("\nBalance Sheet:")
#print(balance_sheet.head())
#print("\nCash Flow Statement:")
#print(cashflow.head())

#Get Historical Prices (Daily or Annual)

# 10+ years of historical data (default interval = 1d)
#history = ticker.history(period="2y")
#history
```

```python
# Extract Revenue and Net Income
revenue = income_stmt.loc["Total Revenue"]
net_income = income_stmt.loc["Net Income"]

# Calculate year-over-year growth (%)
revenue_growth = revenue.pct_change(periods=-1, fill_method=None) * -100
net_income_growth = net_income.pct_change(periods=-1, fill_method=None) * -100

# Show results
print("Revenue Growth (%):")
print(revenue_growth)

print("\nNet Income Growth (%):")
print(net_income_growth)
```

```python
# Extract Revenue and Net Income
revenue = income_stmt.loc["Total Revenue"]
net_income = income_stmt.loc["Net Income"]

# Combine into a DataFrame for display
df = revenue.to_frame(name="Revenue")
df["Net Income"] = net_income

df= pd.DataFrame(df)

print("Available Annual Financial Data:")
df
```
